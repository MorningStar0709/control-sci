# E3 Bonus: Local Fine-Tuning Demo

**Prove** `ControlSci dataset → trainer.train()` with **zero adapter code**.

- **Model**: Qwen2.5-1.5B-Instruct (fp16)
- **GPU**: NVIDIA RTX 5090 Laptop 24GB
- **Data**: 10 samples from full.json
- **Duration**: <15s wall-clock (4.7s pure training)
- **Result**: Loss ↓ 2.25 → 2.05

> After `control-sci-corpus` is published to HF Hub, replace `load_dataset("json", ...)` with:
> `load_dataset("MorningStar0709/control-sci-corpus", "full", split="train")`

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/MorningStar0709/control-sci/blob/main/notebooks/control_sci_demo.ipynb)

```python
# Cell 6: Fine-Tuning Demo — Qwen2.5-1.5B on RTX 5090
# load_dataset → format → tokenize → model.load → trainer.train() = ZERO adapter
import time, json
from datasets import Dataset
from transformers import AutoModelForCausalLM, AutoTokenizer, Trainer, TrainingArguments
import torch

t_start = time.time()

# ── 1. Load Dataset ──
# IF HuggingFace dataset IS published, use:
#   from datasets import load_dataset
#   full = load_dataset("MorningStar0709/control-sci-corpus", "full", split="train")
# ELSE load from local JSON:
with open("benchmark/dataset/full.json", "r", encoding="utf-8") as f:
    raw = json.load(f)
full = Dataset.from_dict({
    "question": [q["question"] for q in raw["questions"]],
    "answer":   [q["answer"]   for q in raw["questions"]],
})

def format_instruction(ex):
    return {"text": f"### 问题\n{ex['question']}\n\n### 参考答案\n{ex['answer']}"}

train_data = full.map(format_instruction).select(range(10))
print(f"✅ Dataset: {len(train_data)} samples")

# ── 2. Tokenizer ──
model_name = "Qwen/Qwen2.5-1.5B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name, padding_side="right")
tokenizer.pad_token = tokenizer.eos_token

def tokenize(ex):
    result = tokenizer(ex["text"], truncation=True, max_length=256)
    result["labels"] = result["input_ids"].copy()
    return result

train_data = train_data.map(tokenize)
print(f"✅ Tokenizer: vocab_size={tokenizer.vocab_size}")

# ── 3. Model ──
model = AutoModelForCausalLM.from_pretrained(model_name, dtype=torch.float16, device_map="auto")
params_b = sum(p.numel() for p in model.parameters()) / 1e9
print(f"✅ Model: {params_b:.1f}B params, device={next(model.parameters()).device}")

# ── 4. Train ──
args = TrainingArguments(
    output_dir="./ctrl_sci_demo_output",
    num_train_epochs=1,
    per_device_train_batch_size=1,
    logging_steps=1,
    report_to="none",
    save_strategy="no",
    learning_rate=2e-5,
    warmup_steps=1,
)

trainer = Trainer(model=model, args=args, train_dataset=train_data)
trainer.train()

# ── 5. Results ──
losses = [e for e in trainer.state.log_history if "loss" in e and e["loss"] < 100]
total_s = time.time() - t_start
print(f"\n✅ load_dataset → trainer.train() 零适配完成 ({total_s:.1f}s)")
print(f"   Loss: {losses[0]['loss']:.4f} → {losses[-1]['loss']:.4f} (↓ 下降, Δ={losses[0]['loss']-losses[-1]['loss']:+.4f})")
print(f"   GPU: {torch.cuda.get_device_name(0)}")
```
